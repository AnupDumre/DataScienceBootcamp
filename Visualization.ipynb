## This is the portion of Pre-Processing.

```python
import pandas as pd
import numpy as np
```

```python
titanic = pd.read_csv('titanic.csv')
titanic.head()
```

```python
titanic.tail()
```

```python
titanic.describe()
```

```python
titanic.info()
```

```python
# Removing the 'Name' column from the Titanic dataset is a reasonable decision as names are unlikely to directly influence a person's survival. Names are usually unique identifiers and do not inherently offer predictive value for survival prediction. However, it's worth noting that certain titles within the name (e.g., "Mr.," "Mrs.," "Miss," etc.) might provide some indication of social status or gender, which could be relevant for survival prediction. Therefore, it is important to perform a thorough data analysis and consider potential implications on the model's performance before proceeding with this deletion. Always validate the decision with data analysis and model evaluation to ensure it doesn't significantly affect the predictive capabilities of the model.

del titanic['Name']
titanic.head()
```

```python
#'Fare' column can never decide survival of a person, hence we can safely delete it

del titanic['Fare']
titanic.head()
```

```python
del titanic['Cabin']
titanic.head()
```

```python
#changing values for "Male, Female" string values to numeric values, i.e, male = 1 and female = 0
def getNumber(str):
        if str == "male":
                return 1
        else:
                return 0
titanic["Gender"]=titanic['Sex'].apply(getNumber)
#We have created new column name "Gender" and filling it with values 1,0 based on the values of sex column
titanic.head()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
titanic['Sex']= labelencoder.fit_transform(titanic['Sex'])
titanic.head()

#this code snippet converts the categorical 'Sex' column in the 'titanic' DataFrame (which contains string values like 'male' and 'female') into numeric values (typically 0 for one category and 1 for the other) using the LabelEncoder from scikit-learn. This numeric encoding allows machine learning algorithms to work with categorical data effectively.
```

###  LabelEncoder uses library so it converts string directly to the boolen.

```python
titanic.isnull().sum() 
#checking for null values in the dataset and summing them up column wise to get the total number of null values in each column
```

```python
#Fill the missing age of the survived one with the mean age of the survival people and non survive age with non survival age ko mean

mean = titanic[titanic.Survived == 1].Age.mean()
mean
```

```python
death = titanic[titanic.Survived == 0].Age.mean()
death
```

```python
# Fill the missing values in the 'age' column of the 'titanic' DataFrame with the 'mean age' of the passengers who survived.
titanic['age'] =np.where(pd.isnull(titanic.Age) & titanic["Survived"]==1 ,mean, titanic['Age'])
titanic
```

```python
titanic.age.fillna(death,inplace=True) # inplace le chai jun column herira ho tei ma operate garne
titanic.head()
```

```python
titanic.age.fillna(death,inplace=True) # inplace le chai jun column herira ho tei ma operate garne
titanic.tail()
```

```python
# now you can see the Age has 177 null but that is replace at 'age' all null value is overcomed
titanic.isnull().sum()
# After filling the missing values, check for the total number of null values in the DataFrame.
```

```python
#To check whether the Embarked will affect the survival of people or not

survivedQ = titanic[titanic.Embarked == 'Q'][titanic.Survived == 1].shape[0]
survivedS = titanic[titanic.Embarked == 'S'][titanic.Survived == 1].shape[0]
survivedC = titanic[titanic.Embarked == 'C'][titanic.Survived == 1].shape[0]
print(survivedQ)
print(survivedS)
print(survivedC)
```

#### The UserWarnings occur because the boolean indexing with multiple conditions is not performed correctly. To fix this, you can use the 'loc' method to perform the boolean indexing with multiple conditions.

```python
# Count the number of survivors for each port of embarkation
survivedQ = titanic.loc[(titanic['Embarked'] == 'Q') & (titanic['Survived'] == 1)].shape[0]
survivedS = titanic.loc[(titanic['Embarked'] == 'S') & (titanic['Survived'] == 1)].shape[0]
survivedC = titanic.loc[(titanic['Embarked'] == 'C') & (titanic['Survived'] == 1)].shape[0]

# Print the results
print(survivedQ)  # Number of survivors from Queenstown (Embarked Q)
print(survivedS)  # Number of survivors from Southampton (Embarked S)
print(survivedC)  # Number of survivors from Cherbourg (Embarked C)
```

```python
# Drop rows with missing values (NaN) from the 'titanic' DataFrame to ensure no missing data remains.

titanic.dropna(inplace=True) #Drop the entire row if it has a null value in any row
titanic.head()
```

```python
titanic.isnull().sum()
```

```python
labelencoder = LabelEncoder()
titanic['Embarked'] = labelencoder.fit_transform(titanic['Embarked'])
titanic.head()
```

```python
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
# drawing a pie chart for the number of males and females

#summing up all the values of column gender with a condition for mle and similary for females
males = (titanic['Sex'] == 1).sum()

females = (titanic['Sex']==0).sum()

print(males)
```

```python
print(females)
```

```python
p = [males,females]
plt.pie(p,                            #giving array
        labels = ['Males','Females'], # coressponding giving Lables
        colors =['green', 'yellow'],  # corresponding colors
        explode = (0.15,0),            #how much the gap should be there between the pies
        startangle = 0)
plt.axis('equal')
plt.show()
```

```python
MaleS = titanic[(titanic.Sex==1)&(titanic.Survived==1)].shape[0]
print(MaleS)

MaleNS = titanic[(titanic.Sex==1)&(titanic.Survived==0)].shape[0]
print(MaleNS)

FemaleS = titanic[(titanic.Sex==0)&(titanic.Survived==1)].shape[0]
print(MaleS)

FemaleNS = titanic[(titanic.Sex==0)&(titanic.Survived==0)].shape[0]
print(FemaleNS)
```

```python
p = [MaleS,MaleNS,FemaleS,FemaleNS]
plt.pie(p,                          #giving array
    labels = ['Male Survived','Male Not-Survived','Female Survived','Female Not-Survived'],     #correspondingly giving lables
    colors=['green','red','blue','yellow'],      # corresponding colors
    explode = (0.015,0.015,0.015,0.015),             #how much the gap should be there between the pies
    counterclock=False,
    autopct='%.2f%%',
    startangle= 0)
plt.axis('equal')
plt.show()
```
